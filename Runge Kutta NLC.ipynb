1st order runge kutta method for NLC. The first order Runge-Kutta method is the same as the Euler's method. In fact, the Euler's method is often referred to as the simplest form of the Runge-Kutta method, where the order of the method is 1.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
T0 = 100  # initial temperature of object (degrees Celsius)
Ts = 20   # temperature of surroundings (degrees Celsius)
k = 0.1   # cooling coefficient

# Time parameters
t0 = 0    # start time
tf = 60   # end time
dt = 1    # time step

# Initialize variables
n = int((tf-t0)/dt) + 1
t = np.linspace(t0, tf, n)
T = np.zeros(n)
T[0] = T0

# First-order Runge-Kutta method
for i in range(n-1):
    k1 = -k*(T[i] - Ts)
    T[i+1] = T[i] + k1*dt

# Plot the results
plt.plot(t, T, label='Temperature')
plt.plot(t, Ts*np.ones(n), 'r--', label='Surrounding temperature')
plt.title("Newton's Law of Cooling")
plt.xlabel('Time (s)')
plt.ylabel('Temperature (Celsius)')
plt.legend()
plt.show()
```

The fourth-order Runge-Kutta method is a numerical algorithm used to approximate the solution of a first-order ordinary differential equation (ODE). The method is based on the Taylor series expansion of the solution function.

The basic idea behind the Runge-Kutta method is to evaluate the slope of the solution function at several points between each pair of grid points, and then to use a weighted average of these slopes to estimate the value of the function at the next grid point. The fourth-order Runge-Kutta method uses four such slope evaluations to compute the weighted average.

At each grid point, the method first computes the slope of the solution function, which is given by the right-hand side of the ODE. This is known as the "k1" term. Next, the method uses this slope to compute an estimate of the solution at the midpoint of the time interval, which is given by the "k2" term. The slope at this midpoint is then used to compute another estimate of the solution at the midpoint, given by the "k3" term. Finally, the method uses the slope at the endpoint of the interval to compute a fourth estimate of the solution, given by the "k4" term.
