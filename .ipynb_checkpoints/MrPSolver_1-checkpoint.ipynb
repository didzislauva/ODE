### Izpletņlēkšanas problēma




Izpletņlēcējs ar masu $m=80kg$ izlec no lidmašīnas, kuras ātrums $v=50m/s$ horizontāli. Ātrums ir vektors ar v_x un v_y vērtībām $ \vec{v}=(v_x, v_y) $

Gaisa pretestības spēks ir dots kā $F_{fr}=-b|\vec{v}|^2\hat{v}$

Smaguma spēks ir dots kā $F_{g}=-mg\hat{h}$

Maksimālais ātrums (terminal velocity) ir $v_t=-55m/s$

1) Atrast funkciju $\vec{v}(t)$, kas ļauj atrast ātrumu jebkurā izvēlētā laika brīdī.

2) Aprēķināt laiku, kas izpletņlēcējam jāpavada gaisā līdz brīdim, kad tiek sasniegts 1% no maksimālā ātruma.

Kad izpletņlēcējs krīt, ir spēkā Ņutona likumi.

2.Ņutona likums vektoriālā formā uzrakstāms kā 
$\vec{F}=
m\vec{a}=
m
\begin{bmatrix}
\frac{dv_x}{dt}\\
\frac{dv_y}{dt}
\end{bmatrix}$

### Divi spēki
#### Gravitācijas spēks
$\vec{F_g}=-m\vec{g}=-m
\begin{bmatrix}
0\\
g
\end{bmatrix}$

#### Gaisa pretestības spēks

$\vec{F_f}=
-b|\vec{v}|\vec{v}=
-b\sqrt{v_x^2+v_y^2}
\begin{bmatrix}
v_x \\
v_y
\end{bmatrix}$



### Kopspēks ir visu spēku summa
$\vec{F}=\vec{F_g}+\vec{F_f}
$


$
m \begin{bmatrix}
\frac{dv_x}{dt} \\
\frac{dv_y}{dt}
\end{bmatrix}
=(-m 
\begin{bmatrix}
0 \\
g
\end{bmatrix})+(
-b\sqrt{v_x^2+v_y^2}
\begin{bmatrix}
v_x \\
v_y
\end{bmatrix})$

No pēdējā vienādojuma var izteikt abu ātrumu atvasinājumu funkcijas - 

$\frac{dv_x}{dt}=-\frac{b}{m}\sqrt{v_x^2+v_y^2}v_x$

un

$\frac{dv_y}{dt}=-\frac{b}{m}\sqrt{v_x^2+v_y^2}v_y-g$

Pārrakstām vēlreiz: 
    
$$
\frac{dv_x}{dt}=-\frac{b}{m}\sqrt{v^2_x+v^2_y}\cdot v_x
\\
\frac{dv_y}{dt}=-g-\frac{b}{m}\sqrt{v^2_x+v^2_y}\cdot v_y

$$

$\vec{S}=(v_x, v_y)$
