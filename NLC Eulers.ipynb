```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parameters
T0 = 100  # initial temperature of object (degrees Celsius)
Ts = 20   # temperature of surroundings (degrees Celsius)
k = 0.1   # cooling coefficient
```

```python
# Time parameters
t0 = 0    # start time
tf = 60   # end time
dt = 1    # time step
```

```python
# Initialize variables
n = int((tf-t0)/dt) + 1 #soļi
t = np.linspace(t0, tf, n) #laika masīvs ar laika vērtību katrā solī
T = np.zeros(n) # aprēķināmo T masīvs vienāds ar nulli. Vēlāk saliks iekšā. 
T[0] = T0 #nulltā  T masīva vērtība vienāda ar sākuma nosacījumu. Pārējos risinās ciklā
```

```python
# Euler's method
for i in range(n-1):
    T[i+1] = T[i] + (-k*(T[i] - Ts))*dt 
    
#Eilera metode tā vietā, lai rakstītu analītisko, T(t)=T_apk+(T0-T_apk)e^-kt
```

```python
plt.plot(t, T, label='Temperature')
plt.plot(t, Ts*np.ones(n), 'r--', label='Surrounding temperature')
plt.title("Newton's Law of Cooling")
plt.xlabel('Time (s)')
plt.ylabel('Temperature (Celsius)')
plt.legend()
plt.show()
```

```python
# Parameters
T_0 = 100.0    # Initial temperature of object in Celsius
T_s = 20.0     # Temperature of surroundings in Celsius
k = 0.1        # Cooling coefficient of the object

# Simulation time
t_max = 300.0   # Maximum simulation time in seconds
delta_t = 1.0   # Time step size in seconds
num_steps = int(t_max / delta_t)

# Arrays to store temperature and time values
T = np.zeros(num_steps + 1)
T[0] = T_0
t = np.linspace(0.0, t_max, num_steps + 1)

# Euler's method
for i in range(num_steps):
    dT_dt = -k * (T[i] - T_s)
    T[i+1] = T[i] + dT_dt * delta_t

# Plot results
plt.plot(t, T)
plt.xlabel('Time (s)')
plt.ylabel('Temperature (Celsius)')
plt.title('Cooling of Object')
plt.show()
```

Bet ja apkārtējās vides temperatūra pieaug?

```python
# Define parameters
T0 = 100  # initial temperature of object
T_surroundings_0 = 20  # initial temperature of surroundings
k = 0.1  # cooling coefficient
h = 0.001  # time step (in seconds)
t_max = 1000  # maximum time (in seconds)
T_surroundings = lambda t: 20 + 0.1 * t  # temperature of surroundings as a function of time

# Define lists to store results
time = [0]
T_object = [T0]
T_surroundings_list = [T_surroundings(0)]

# Apply Euler's method to numerically solve the differential equation
for i in range(int(t_max/h)):
    T_object.append(T_object[-1] - k*(T_object[-1]-T_surroundings_list[-1])*h)
    time.append(time[-1] + h)
    T_surroundings_list.append(T_surroundings(time[-1]))

# Plot the results
plt.plot(time, T_object, label='Object')
plt.plot(time, T_surroundings_list, label='Surroundings')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (C)')
plt.legend()
plt.show()
```

un tagad apkārtējās vides temperatūras pieaugums ir atkarīgs no objekta temperatūras. Otrais termodinamikas likums.

```python
# Define parameters
T_obj = 100    # Object initial temperature in C
T_sur = 20     # Surrounding initial temperature in C
k1 = 0.1        # Cooling coefficient
k2 = 0.01      # sasilsanas koeficients

# Define time step and simulation time
dt = 0.1       # Time step in seconds
t_max = 60     # Simulation time in seconds

# Create lists to store results
T_obj_list = [T_obj]
T_sur_list = [T_sur]
t_list = [0]

# Simulate cooling process
for i in range(int(t_max/dt)):
    # Calculate change in temperatures
    dT_obj = -k1*(T_obj - T_sur)*dt
    dT_sur = (k2*(T_obj - T_sur) - (T_sur - 20)*0.0001)*dt

    # Update temperatures
    T_obj += dT_obj
    T_sur += dT_sur

    # Store results
    T_obj_list.append(T_obj)
    T_sur_list.append(T_sur)
    t_list.append((i+1)*dt)

# Plot results
plt.plot(t_list, T_obj_list, label='Object temperature')
plt.plot(t_list, T_sur_list, label='Surrounding temperature')
plt.xlabel('Time (s)')
plt.ylabel('Temperature (C)')
plt.title('Temperature vs. time')
plt.legend()
plt.show()
```
